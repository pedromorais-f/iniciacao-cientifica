```python
from functions import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Plot Function
def plot_random_numbers(data_list_x, data_list_y):
    #figure size
    plt.figure(figsize= (25, 10))
    
    #Format
    plt.scatter(data_list_x, data_list_y)
    
    #Informations
    plt.title("Random Numbers")
    plt.xlabel("X Numbers")
    plt.ylabel("Y Numbers")
    
    #Customization
    plt.grid(True)
    plt.box(False)
```

```python
#Getting the values from txt file
points1, points2 = read_txt()
print(points1)
print(points2)
```

```python
#Plotting the numbers
plot_random_numbers(points1, points2)
```

```python
#Getting the values from csv file
points1, points2 = read_csv()
print(points1)
print(points2)
```

```python
#Plotting the numbers
plot_random_numbers(points1, points2)
```
